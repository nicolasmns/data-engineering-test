```python
from src.task_2_contact_fullname import *
import matplotlib.pyplot as plt
from src.utils import getFilepath
import pandas as pd
import ast
import os


orders_filepath = getFilepath("orders.csv")
```

```python
df_1 = create_orders_with_contact_names(load_csv_to_dataframe(orders_filepath))
df_1.head()
```

```python

contact_counts = df_1["contact_full_name"].value_counts()

plt.figure(figsize=(10, 6))
contact_counts.plot(kind='bar', color='skyblue')
plt.title("Number of Orders per Contact", fontsize=16)
plt.xlabel("Contact Name", fontsize=12)
plt.ylabel("Number of Orders", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
missing_contacts = (df_1["contact_full_name"] == "John Doe").sum()
valid_contacts = len(df_1) - missing_contacts

plt.figure(figsize=(8, 8))
plt.pie([valid_contacts, missing_contacts],
        labels=["Valid Contacts", "Missing Contacts"],
        autopct='%1.1f%%', startangle=140, colors=['#66c2a5', '#fc8d62'])
plt.title("Proportion of Valid vs Missing Contacts", fontsize=16)
plt.show()
```
