```python
from src.task_5_sales_owners import *
import matplotlib.pyplot as plt
```

```python
df3 = generate_list_sales_owners()
df3.head()
```

```python
df3["num_salesowners"] = df3["list_salesowners"].apply(lambda x: len(x.split(", ")))

# Plot
plt.figure(figsize=(10, 6))
df_sorted = df3.sort_values(by="num_salesowners", ascending=False)
plt.bar(df_sorted["company_name"], df_sorted["num_salesowners"], color="skyblue")
plt.title("Number of Sales Owners per Company", fontsize=16)
plt.xlabel("Company Name", fontsize=12)
plt.ylabel("Number of Sales Owners", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
df3["num_salesowners"] = df3["list_salesowners"].apply(lambda x: len(x.split(", ")))

# Group by the number of sales owners
salesowners_distribution = df3["num_salesowners"].value_counts()

# Plot
plt.figure(figsize=(8, 8))
plt.pie(
    salesowners_distribution,
    labels=salesowners_distribution.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors
)
plt.title("Distribution of Companies by Number of Sales Owners", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
salesowners = df3["list_salesowners"].str.split(", ").explode()

# Count occurrences of each sales owner
salesowner_counts = salesowners.value_counts()

# Plot
plt.figure(figsize=(12, 6))
salesowner_counts.plot(kind="bar", color="orange")
plt.title("Frequency of Sales Owners Across Companies", fontsize=16)
plt.xlabel("Sales Owner", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
